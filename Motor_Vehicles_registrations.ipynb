## 19. Query Planning Layer Using Gemini

Converting natural language questions into structured execution plans.

````python
import json

def get_query_plan_gemini(question):

    prompt = f"""
You are a data query planner.

Dataset Schema:

State : Name of State or UT
Year : 2006-2016
Vehicles : Registered Motor Vehicles

Your job:

1. Determine whether the question can be answered from this dataset.
2. If answerable, generate a JSON plan.
3. If not answerable, return out_of_scope.

Allowed operations:

rank
trend
compare
aggregate
growth

Examples:

Question:
Which state had the highest vehicles in 2016?

Output:
{{
"answerable": true,
"operation": "rank",
"year": 2016,
"order": "desc",
"limit": 1
}}

Question:
Compare Maharashtra and Gujarat

Output:
{{
"answerable": true,
"operation": "compare",
"states": ["Maharashtra","Gujarat"]
}}

Question:
Show trend of Kerala

Output:
{{
"answerable": true,
"operation": "trend",
"state": "Kerala"
}}

Question:
What was Delhi AQI in 2016?

Output:
{{
"answerable": false,
"reason": "Dataset contains vehicle registration data only."
}}

Question:
{question}

Return ONLY JSON.
"""

    print("Calling Gemini...")

    response = model.generate_content(prompt)

    print("Gemini responded")

    text = response.text.strip()

    print("Response text:")
    print(text)

    if text.startswith("```"):
        text = text.replace("```json", "")
        text = text.replace("```", "")
        text = text.strip()

    return json.loads(text)
````

#20. Groq Query Planner

````python
import json

def get_query_plan_groq(question):

    prompt = f"""
You are a query planner for a vehicle registration dataset.

Dataset:
- State
- Year (2006-2016)
- Vehicles

Return ONLY valid JSON.

Examples:

Question:
Which state had the highest registered vehicles in 2016?

Output:
{{
    "answerable": true,
    "operation": "rank",
    "year": 2016,
    "order": "desc",
    "limit": 1
}}

Question:
Top 5 states by registered vehicles in 2016.

Output:
{{
    "answerable": true,
    "operation": "rank",
    "year": 2016,
    "order": "desc",
    "limit": 5
}}

Question:
Show trend of Bihar.

Output:
{{
    "answerable": true,
    "operation": "trend",
    "state": "Bihar"
}}

Question:
Compare Maharashtra and Gujarat.

Output:
{{
    "answerable": true,
    "operation": "compare",
    "states": ["Maharashtra", "Gujarat"]
}}

Question:
What was the total number of registered vehicles in India in 2012?

Output:
{{
    "answerable": true,
    "operation": "aggregate",
    "year": 2012
}}

Question:
What was Delhi AQI in 2016?

Output:
{{
    "answerable": false,
    "reason": "Dataset contains vehicle registration data only."
}}

Now answer this question:

{question}

Return ONLY JSON.
"""

    response = groq_client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    text = response.choices[0].message.content.strip()

    if text.startswith("```"):
        text = text.replace("```json", "")
        text = text.replace("```", "")
        text = text.strip()

    return json.loads(text)
````

# 22. Query Execution Engine

Executing analytical operations on the dataset using Pandas.

````python
def execute_query(plan):

    if not plan["answerable"]:
        return {
            "type": "out_of_scope",
            "answer": plan["reason"]
        }

    operation = plan["operation"]

    if operation == "rank":

        filtered = df_long[
            df_long["Year"] == plan["year"]
        ]

        if "states" in plan:

            filtered = filtered[
                filtered["State"].isin(
                    plan["states"]
                )
            ]

        result = (
            filtered
            .sort_values(
                "Vehicles",
                ascending=(plan["order"] == "asc")
            )
            .head(plan["limit"])
        )

        return {
            "type": "table",
            "data": result
        }

    elif operation == "trend":

        states = plan.get(
            "states",
            [plan.get("state")]
        )

        start_year = plan.get(
            "start_year",
            2006
        )

        end_year = plan.get(
            "end_year",
            2016
        )

        result = (
            df_long[
                (df_long["State"].isin(states))
                &
                (df_long["Year"] >= start_year)
                &
                (df_long["Year"] <= end_year)
            ]
            .sort_values(["State", "Year"])
        )

        return {
            "type": "trend",
            "data": result
        }


    elif operation == "compare":

        year = plan.get("year", 2016)

        if plan.get("over_years", False):

            result = (
                df_long[
                    (df_long["State"].isin(plan["states"]))
                    &
                    (df_long["Year"] <= year)
                ]
                .sort_values(
                    ["State", "Year"]
                )
            )

        else:

            result = (
                df_long[
                    (df_long["State"].isin(plan["states"]))
                    &
                    (df_long["Year"] == year)
                ]
                .sort_values(
                    "Vehicles",
                    ascending=False
                )
            )

        return {
            "type": "compare",
            "data": result,
            "year": year,
            "over_years": plan.get(
                "over_years",
                False
            )
        }


    elif operation == "compare_all":

        result = (
            df_long[
                df_long["Year"] == plan["year"]
            ]
            .sort_values(
                "Vehicles",
                ascending=False
            )
        )

        if plan.get("limit"):

            result = result.head(
                plan["limit"]
            )

            target_state = plan["state"]

            if (
                target_state
                not in result["State"].values
            ):

                target_row = (
                    df_long[
                        (df_long["State"] == target_state)
                        &
                        (df_long["Year"] == plan["year"])
                    ]
                )

                result = pd.concat(
                    [
                        result,
                        target_row
                    ]
                )

        return {
            "type": "compare_all",
            "data": result,
            "state": plan["state"],
            "year": plan["year"]
        }

    elif operation == "aggregate":

        total = (
            df_long[
                df_long["Year"] == plan["year"]
            ]["Vehicles"]
            .sum()
        )

        return {
            "type": "number",
            "value": total
        }


    elif operation == "state_year":

        result = (
            df_long[
                (df_long["State"] == plan["state"])
                &
                (df_long["Year"] == plan["year"])
            ]
        )

        return {
            "type": "state_year",
            "data": result
        }


    elif operation == "growth":

        growth_df = []

        for state in df_long["State"].unique():

            state_data = (
                df_long[
                    df_long["State"] == state
                ]
                .sort_values("Year")
            )

            start = state_data.iloc[0]["Vehicles"]
            end = state_data.iloc[-1]["Vehicles"]

            growth = end - start

            growth_df.append(
                [state, growth]
            )

        growth_df = pd.DataFrame(
            growth_df,
            columns=["State", "Growth"]
        )

        result = (
            growth_df
            .sort_values(
                "Growth",
                ascending=False
            )
            .head(10)
        )

        return {
            "type": "growth",
            "data": result
        }
````

## 23. Local Answer Generation

Generating responses directly from computed results without additional LLM calls.

````python
def generate_answer_local(question, result):



    if result["type"] == "out_of_scope":
        return result["answer"]

    # Ranking Results
    if result["type"] == "table":

        df = result["data"]

        if len(df) == 1:

            row = df.iloc[0]

            return (
                f"{row['State']} had "
                f"{row['Vehicles']:.2f} "
                f"registered vehicles."
            )

        lines = []

        for i, (_, row) in enumerate(df.iterrows(), start=1):

            lines.append(
                f"{i}. {row['State']} : {row['Vehicles']:.2f}"
            )

        return "\n".join(lines)

    # Aggregate Results
    if result["type"] == "number":

        return (
            f"Total registered vehicles: "
            f"{result['value']:.2f}"
        )

    # Trend Results
    if result["type"] == "trend":

        states = result["data"]["State"].unique()

        if len(states) == 1:

            state = states[0]

            state_data = result["data"]

            start_year = state_data.iloc[0]["Year"]
            end_year = state_data.iloc[-1]["Year"]

            start = state_data.iloc[0]["Vehicles"]
            end = state_data.iloc[-1]["Vehicles"]

            growth = end - start

            return (
                f"{state} increased from "
                f"{start:.2f} vehicles in {start_year} "
                f"to {end:.2f} vehicles in {end_year}. "
                f"Net growth: {growth:.2f} vehicles."
            )

        lines = []

        for state in states:

            state_data = result["data"][
                result["data"]["State"] == state
            ]

            start_year = state_data.iloc[0]["Year"]
            end_year = state_data.iloc[-1]["Year"]

            start = state_data.iloc[0]["Vehicles"]
            end = state_data.iloc[-1]["Vehicles"]

            lines.append(
                f"{state}: {start:.2f} ({start_year}) → "
                f"{end:.2f} ({end_year})"
            )
        return "\n".join(lines)

    # Compare Results
    if result["type"] == "compare":

        if result.get("over_years", False):

            states = result["data"]["State"].unique()

            lines = []

            for state in states:

                state_data = result["data"][
                    result["data"]["State"] == state
                ]

                start = state_data.iloc[0]["Vehicles"]
                end = state_data.iloc[-1]["Vehicles"]

                lines.append(
                    f"{state}: "
                    f"{start:.2f} ({state_data.iloc[0]['Year']}) → "
                    f"{end:.2f} ({state_data.iloc[-1]['Year']})"
                  )

            return (
                "Comparison over years:\n"
                +
                "\n".join(lines)
            )

        latest = result["data"]

        latest = latest.sort_values(
            "Vehicles",
            ascending=False
        )

        lines = []

        for _, row in latest.iterrows():

            lines.append(
                f"{row['State']} : "
                f"{row['Vehicles']:.2f}"
            )

        return (
            f"Vehicle registrations in {result['year']}:\n"
            + "\n".join(lines)
        )

    elif result["type"] == "compare_all":

        state = result["state"]

        full_df = (
            df_long[
                df_long["Year"] == result["year"]
            ]
            .sort_values(
                "Vehicles",
                ascending=False
            )
            .reset_index(drop=True)
        )

        if state in full_df["State"].values:

            rank = (
                full_df[
                    full_df["State"] == state
                ]
                .index[0]
                + 1
            )

            vehicles = (
                full_df[
                    full_df["State"] == state
                ]["Vehicles"]
                .iloc[0]
            )

            return (
                f"In {result['year']}, "
                f"{state} ranked #{rank} "
                f"among all states "
                f"with {vehicles:.2f} "
                f"registered vehicles."
            )


    # State-Year Lookup Results
    if result["type"] == "state_year":

        if len(result["data"]) == 0:
            return "No data found."

        row = result["data"].iloc[0]

        return (
            f"{row['State']} had "
            f"{row['Vehicles']:.2f} "
            f"registered vehicles in "
            f"{int(row['Year'])}."
        )

    # Growth Results
    if result["type"] == "growth":

        top = result["data"].iloc[0]

        return (
            f"{top['State']} experienced the "
            f"highest growth of "
            f"{top['Growth']:.2f} vehicles "
            f"between 2006 and 2016."
        )

    return "Computation completed."
````

# 24. Visualization Engine

Creating charts for trends, comparisons, rankings and growth analysis.

````python
import plotly.express as px

def create_chart(result):

    if result["type"] == "trend":

        fig = px.line(
            result["data"],
            x="Year",
            y="Vehicles",
            color="State",
            title="Vehicle Registration Trend",
            markers=True
        )

        return fig

    elif result["type"] == "compare":

        if result.get("over_years", False):

            fig = px.line(
                result["data"],
                x="Year",
                y="Vehicles",
                color="State",
                title="State Comparison Over Years",
                markers=True
            )

        else:

            fig = px.bar(
                result["data"],
                x="State",
                y="Vehicles",
                color="State",
                title=f"State Comparison ({result['year']})"
            )

        return fig

    elif result["type"] == "compare_all":

        fig = px.bar(
            result["data"],
            x="Vehicles",
            y="State",
            orientation="h",
            title=f"Comparison ({result['year']})"
        )

        target_state = result["state"]

        fig.update_traces(
            marker_color=[
                "crimson" if s == target_state else "blue"
                for s in result["data"]["State"]
            ]
        )

        return fig

    elif result["type"] == "table":

        fig = px.bar(
            result["data"],
            x="State",
            y="Vehicles",
            title="Vehicle Registrations"
        )

        return fig

    elif result["type"] == "growth":

        fig = px.bar(
            result["data"],
            x="State",
            y="Growth",
            title="Growth in Vehicle Registrations (2006-2016)"
        )

        return fig

    elif result["type"] == "state_year":

        fig = px.bar(
            result["data"],
            x="State",
            y="Vehicles",
            title="Registered Vehicles"
        )

        return fig

    elif result["type"] == "number":

        fig = px.bar(
            x=["Total"],
            y=[result["value"]],
            title="Total Registered Vehicles"
        )

        return fig

    return None
````

Gemini is used only for intent extraction and query planning.

All filtering, aggregation, ranking, comparison, trend analysis and growth computations are executed locally using Pandas. This minimizes API usage, improves reproducibility and ensures that numerical results originate from dataset computation rather than model memory.

Every answer follows the pipeline:

User Question → Query Plan → Pandas Execution → Answer Generation

````python
import gradio as gr
import re

# -----------------------------
# Fast Local Planner
# -----------------------------
def local_query_plan(question):

    q = question.lower().strip()

    year_match = re.search(r"(20\d\d)", q)

    if year_match:

        year = int(year_match.group(1))

        if year < 2006 or year > 2016:

            return {
                "answerable": False,
                "reason": "Dataset contains vehicle registration data only for years 2006 to 2016."
            }



      # Telangana validation

    if "telangana" in q and year_match:

        if int(year_match.group(1)) < 2014:

            return {
                "answerable": False,
                "reason": (
                    "Telangana was formed in 2014. "
                    "No Telangana vehicle registration data exists before 2014."
                )
            }


    out_keywords = [
        "aqi",
        "gdp",
        "wheat",
        "rainfall",
        "temperature",
        "population",
        "accident"
    ]

    if any(word in q for word in out_keywords):

        return {
            "answerable": False,
            "reason": "Dataset contains vehicle registration data only."
        }


    # Trend

    if "trend" in q:

        states = []

        q_clean = (
            q.lower()
            .replace("&", "and")
            .replace(",", "")
        )

        for state in df_long["State"].unique():

            state_clean = (
                state.lower()
                .replace("&", "and")
                .replace(",", "")
            )

            if state_clean in q_clean:

                states.append(state)

        years = re.findall(r"(20\d\d)", q)

        if len(years) == 1:

            return {
                "answerable": False,
                "reason": (
                    "Trend analysis requires a year range. "
                    "Please specify both start and end years, "
                    "for example: Show trend of Bihar from 2014 to 2016."
                )
            }

        start_year = 2006
        end_year = 2016

        if len(years) >= 2:

            start_year = int(years[0])
            end_year = int(years[1])

        if len(states) >= 1:

            return {
                "answerable": True,
                "operation": "trend",
                "states": states,
                "start_year": start_year,
                "end_year": end_year
            }


    # Compare



    if "compare" in q:

        states = []

        q_clean = (
            q.lower()
            .replace("&", "and")
            .replace(",", "")
        )

        for state in df_long["State"].unique():

            state_clean = (
                state.lower()
                .replace("&", "and")
                .replace(",", "")
          )

            if state_clean in q_clean:

                states.append(state)

        if len(states) >= 2:

            year = 2016

            if year_match:
                year = int(year_match.group(1))


            over_years = (
                "over years" in q
                or
                "across years" in q
                or
                "throughout years" in q
            )

            return {
                "answerable": True,
                "operation": "compare",
                "states": states,
                "year": year,
                "over_years": over_years
            }



        if len(states) == 1:

            top_match = re.search(
                r"top\s+(\d+)",
                q
            )

            if (
                "other states" in q
                or
                "all states" in q
                or
                "other state" in q
                or
                "all state" in q
                or
                top_match
            ):

                year = 2016

                if year_match:
                    year = int(
                        year_match.group(1)
                    )

                limit = None

                if top_match:
                    limit = int(
                        top_match.group(1)
                    )

                return {
                    "answerable": True,
                    "operation": "compare_all",
                    "state": states[0],
                    "year": year,
                    "limit": limit
                }


    # Growth
    if any(word in q for word in [
        "growth",
        "increase",
        "expansion",
        "rise",
        "grew",
        "growth rate",
        "highest growth",
        "largest increase",
        "fastest growth",
        "most growth"
    ]):

        return {
            "answerable": True,
            "operation": "growth"
        }

    # Aggregate
    if "total" in q:

        year = 2016

        if year_match:
            year = int(year_match.group(1))

        return {
            "answerable": True,
            "operation": "aggregate",
            "year": year
        }



    # Ranking


    year = 2016

    if year_match:
        year = int(year_match.group(1))

    limit = 1

    top_match = re.search(r"top\s+(\d+)", q)

    if top_match:
        limit = int(top_match.group(1))

    if "highest" in q or "top" in q:

        return {
            "answerable": True,
            "operation": "rank",
            "year": year,
            "order": "desc",
            "limit": limit
        }

    if "lowest" in q:

        return {
            "answerable": True,
            "operation": "rank",
            "year": year,
            "order": "asc",
            "limit": limit
        }



    ranking_words = [
    "which",
    "best",
    "highest",
    "top",
    "strongest",
    "performer",
    "rank",
    "among",
    "dominated",
    "leader"
    ]

    if any(word in q for word in ranking_words):
        return None

    # State + Year Lookup


    q_clean = (
        q.lower()
        .replace("&", "and")
        .replace(",", "")
    )

    for state in df_long["State"].unique():

        state_clean = (
            state.lower()
            .replace("&", "and")
            .replace(",", "")
        )

        if state_clean in q_clean:

            year = 2016

            if year_match:
                year = int(year_match.group(1))

            return {
                "answerable": True,
                "operation": "state_year",
                "state": state,
                "year": year
            }

    return None




# -----------------------------
# Gemini Cache
# -----------------------------
query_cache = {}

def get_cached_plan(question):

    if question in query_cache:

        return query_cache[question]

    plan = get_query_plan(question)

    query_cache[question] = plan

    return plan



# -----------------------------
# Main Function
# -----------------------------





def ask_question(question):

    if not question.strip():

        return (
            "Please enter a question.",
            None
        )

    try:

        plan = local_query_plan(question)

        if plan is None:

            q = question.lower()

            dataset_keywords = [
                "vehicle",
                "vehicles",
                "registration",
                "registered",
                "state",
                "states",
                "year",
                "trend",
                "compare",
                "growth",
                "increase",
                "highest",
                "lowest",
                "top",
                "total"
            ]

            contains_state = any(
                state.lower() in q
                for state in df_long["State"].unique()
            )

            if (
                not any(
                    word in q
                    for word in dataset_keywords
                )
                and
                not contains_state
            ):

                return (
                    "Out of scope. This system only answers questions about registered motor vehicles in India (2006–2016).",
                    None
                )

            plan = get_cached_plan(question)

        plan = normalize_plan(plan)
        print("QUESTION:", question)
        print("PLAN:", plan)

        result = execute_query(plan)

        answer = generate_answer_local(
            question,
            result
        )

        fig = create_chart(result)

        return answer, fig

    except Exception as e:

        return (
            f"Error: {str(e)}",
            None
        )


# -----------------------------
# UI
# -----------------------------
with gr.Blocks(
    title="Vehicle Registration Analytics Assistant"
) as demo:

    gr.Markdown(
        """
# 🚗 Vehicle Registration Analytics Assistant

Ask natural language questions about State/UT-wise Registered Motor Vehicles in India (2006–2016).

### Supported Analytics
- Ranking
- Top-N Analysis
- Trend Analysis
- State Comparison
- Aggregation
- Growth Analysis
- Out-of-Scope Detection
"""
    )

    question_box = gr.Textbox(
        label="Ask a Question",
        placeholder="Example: Show trend of Bihar",
        lines=3
    )

    submit_btn = gr.Button(
        "Generate Answer",
        variant="primary"
    )

    answer_box = gr.Textbox(
        label="Answer",
        lines=12,
        max_lines=20
    )

    chart_output = gr.Plot(
        label="Visualization"
    )

    gr.Examples(
        examples=[
            ["Which state had the highest registered vehicles in 2016?"],
            ["Top 5 states by registered vehicles in 2016."],
            ["Show trend of Bihar."],
            ["Compare Maharashtra and Gujarat."],
            ["What was the total number of registered vehicles in India in 2012?"],
            ["Which state experienced the highest growth in vehicle registrations between 2006 and 2016?"],
            ["What was Delhi AQI in 2016?"]
        ],
        inputs=question_box
    )

    submit_btn.click(
        fn=ask_question,
        inputs=question_box,
        outputs=[
            answer_box,
            chart_output
        ]
    )

demo.launch(
    share=True,
    debug=False
)
````
